```
import numpy as np
import cvxpy 
# Exchange rate data.
tickers = ["USD", "EUR", "GBP", "CAD", "JPY", "CNY", "RUB", "MXN", "INR", "BRL"]
n = len(tickers)
shape = (n,n)
F = np.zeros(shape)
X = cvxpy.Variable(shape)
# USD
data = ([1.0, 0.87, 0.76, 1.31, 108.90, 6.72, 65.45, 19.11, 71.13, 3.69],
# EUR
[1.0, 0.88, 1.51, 125.15, 7.72, 75.23, 21.96, 81.85, 4.24],
# GBP
[1.0, 1.72, 142.94, 8.82, 85.90, 25.08, 93.50, 4.84],
# CAD
[1.0, 82.93, 5.11, 49.82, 14.54, 54.23, 2.81],
# JPY
[1.0, 0.062, 0.60, 0.18, 0.65, 0.034],
# CNY
[1.0, 9.74, 2.85, 10.61, 0.55],
# RUB
[1.0, 0.29, 1.09, 0.056],
# MXN
[1.0, 3.73, 0.19],
# INR
[1.0, 0.052],
# BRL
[1.0])
for i in range(n):
    F[i,i:] = data[i]
for j in range(n):
    for i in range(j+1,n):
        F[i,j] = 1.035/F[j,i]
        
# Initial and final portfolios.
c_req = np.arange(1,n+1)
c_req = 1e4*c_req/c_req.sum()
c_init = c_req[::-1]


X_ex = X/F
temp = X.T@np.ones(n) - X_ex@np.ones(n)
c =  c_init - temp
cost = temp@np.sqrt(F[:,0]/F[0,:])
objective = cvxpy.Minimize(cost)
constraints = [X >= 0,c >= c_req,X.T@np.ones(n) <= c_init,cvxpy.diag(X) == 0]
P = cvxpy.Problem(objective,constraints)
X_opt = P.solve()
print("Optimal X value : ", X_opt)
```
