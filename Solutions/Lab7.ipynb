```
L_c = cp.square(c[N-1]-c[0])
for i in range(N-1):
  L_c += cp.sum((c[i]-c[i+1])**2)
```

```
objective = cp.Minimize(L_c+1*(np.ones(T)@s))
```

```
r_abs = cp.norm(r,1)/T
constraints = [ c>=0,s>=0,s<=c,p==(c-s+r),r_abs <=4]
for i in range(N,T):
    constraints += [c[i] == c[i-N]]
```

```
P = cp.Problem(objective,constraints)
P.solve()
```
